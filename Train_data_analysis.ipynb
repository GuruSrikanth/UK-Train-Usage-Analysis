```python
file_path = 'output.xlsx'  # Update this to your file path
df = pd.read_excel(file_path)
yearly_trends = df.groupby(df['Year'])['Current year\nEntries & Exits'].sum()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# # Convert 'Year' to datetime if needed (or you can just use the year directly if that's more appropriate)
# df['Year'] = pd.to_datetime(df['Year'].str[:4], format='%Y')

# Calculate year-over-year growth for each station
# This can be done by pivoting the table and calculating the percentage change year over year
df_pivot = df.pivot_table(index='Station Name', columns=df['Year'], values='Current year\nEntries & Exits')

# Calculate the growth from the first year available to the last year available for each station
# Assuming the DataFrame has sorted columns by year
growth = df_pivot.iloc[:, -1] / df_pivot.iloc[:, 0] - 1

# Sort the stations by growth
sorted_growth = growth.sort_values(ascending=False)

# Plotting the top 10 stations by growth
sorted_growth.head(10).plot(kind='bar', figsize=(10, 6))
plt.title('Top 10 Stations by Growth')
plt.xlabel('Station Name')
plt.ylabel('Growth Rate')
plt.show()
```

```python
from scipy import stats

df['z_score'] = stats.zscore(df['Current year\nEntries & Exits'])
# Assuming a threshold of 3 for anomaly detection
df_anomalies = df[abs(df['z_score']) > 3]
df_anomalies
```

```python
Q1 = df['Current year\nEntries & Exits'].quantile(0.25)
Q3 = df['Current year\nEntries & Exits'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_anomalies = df[(df['Current year\nEntries & Exits'] < lower_bound) | (df['Current year\nEntries & Exits'] > upper_bound)]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['Current year\nEntries & Exits'], label='Year')
plt.scatter(df_anomalies['Year'], df_anomalies['Current year\nEntries & Exits'], color='red', label='Anomaly')
plt.title('Anomaly Detection in Station Usage')
plt.xlabel('Year')
plt.ylabel('Current year\nEntries & Exits')
plt.legend()
plt.show()
```

```python
grouped = df.groupby(['Station Facility Owner', df['Year']])['Current year\nEntries & Exits'].sum().reset_index()
grouped
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style('whitegrid')

# Create a line plot for each facility owner
plt.figure(figsize=(14, 7))
sns.lineplot(data=grouped, x='Year', y='Current year\nEntries & Exits', hue='Station Facility Owner')
plt.title('Usage Trends by Station Facility Owner')
plt.xlabel('Year')
plt.ylabel('Total Entries & Exits')
plt.legend(title='Facility Owner', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import f_oneway

# Example: ANOVA test for three different facility owners
owner1 = df[df['Station Facility Owner'] == 'London Underground']['Current year\nEntries & Exits']
owner2 = df[df['Station Facility Owner'] == 'London Overground']['Current year\nEntries & Exits']
owner3 = df[df['Station Facility Owner'] == 'c2c']['Current year\nEntries & Exits']
owner3 = df[df['Station Facility Owner'] == 'Chiltern Railways']['Current year\nEntries & Exits']

f_stat, p_val = f_oneway(owner1, owner2, owner3)
print('F-Statistic:', f_stat, 'P-Value:', p_val)
```
